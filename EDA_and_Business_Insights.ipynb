# Importing Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the datasets

```python
customers_df = pd.read_csv('Customers.csv')
products_df = pd.read_csv('Products.csv')
transactions_df = pd.read_csv('Transactions.csv')
```

# Convert date columns to datetime

```python
customers_df['SignupDate'] = pd.to_datetime(customers_df['SignupDate'])
transactions_df['TransactionDate'] = pd.to_datetime(transactions_df['TransactionDate'])
```

# --- EDA and Insights ---
# 1. Region Distribution of Customers

```python
region_distribution = customers_df['Region'].value_counts()
print("Region Distribution:")
print(region_distribution)
```

# Visualization

```python
plt.figure(figsize=(8, 5))
region_distribution.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Region Distribution of Customers', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('region_distribution.png')
plt.show()
```

# Insight 1

```python
print("Insight 1: South America has the highest customer base, contributing significantly to overall sales.")
```

# 2. Product Category Distribution

```python
product_category_distribution = products_df['Category'].value_counts()
print("\nProduct Category Distribution:")
print(product_category_distribution)
```

# Visualization

```python
plt.figure(figsize=(8, 5))
product_category_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'orange', 'lightgreen', 'pink'], textprops={'fontsize': 12})
plt.title('Product Category Distribution', fontsize=14)
plt.ylabel('')  # Hide y-axis label for pie chart
plt.tight_layout()
plt.savefig('product_category_distribution.png')
plt.show()
```

# Insight 2

```python
print("Insight 2: Books and Electronics dominate the product categories, each contributing 26% to the catalog.")
```

# 3. Total Transactions and Revenue

```python
total_transactions = transactions_df['TransactionID'].nunique()
total_revenue = transactions_df['TotalValue'].sum()
avg_revenue_per_transaction = transactions_df['TotalValue'].mean()

print("\nTotal Transactions:", total_transactions)
print("Total Revenue: $", total_revenue)
print("Average Revenue Per Transaction: $", avg_revenue_per_transaction)
```

# Insight 3

```python
print("Insight 3: The company has generated $689,995.56 in revenue across 1,000 transactions, averaging $690 per transaction.")
```

# 4. Most Purchased Product

```python
most_purchased_product = transactions_df.groupby('ProductID')['Quantity'].sum().idxmax()
most_purchased_product_name = products_df.loc[products_df['ProductID'] == most_purchased_product, 'ProductName'].values[0]
print("\nMost Purchased Product:", most_purchased_product_name)
```

# Insight 4

```python
print("Insight 4: The 'SoundWave Cookbook' is the most purchased product, indicating high customer interest in this item.")
```

# 5. Customer Signup Trend

```python
customers_df['SignupYear'] = customers_df['SignupDate'].dt.year
signup_trend = customers_df['SignupYear'].value_counts().sort_index()
```

# Visualization

```python
plt.figure(figsize=(8, 5))
signup_trend.plot(kind='line', marker='o', color='green')
plt.title('Customer Signup Trend Over the Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Signups', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('customer_signup_trend.png')
plt.show()
```

# Insight 5

```python
print("Insight 5: Customer signups peaked in 2022, highlighting successful marketing strategies during that period.")
```
